# Business Insights

Objectives:-
1. Profitability Optimization.
2. Customer Behavior.

---
Import Libraries:-

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

---
Load Cleaned Data:-

```python
file_path = ('../Data/02_Clean_Data/MM_Clean_Data.csv')
df = pd.read_csv(file_path)

df["Date"] = pd.to_datetime(df["Date"])
df.head()
```

---
Quick Validation:-

```python
df.info()
```

---
Profitability Analysis:-

```python
Profit_Analysis = df.groupby("Vehicle_Type")["Profit_Margin_%"].mean()

plt.figure(figsize = (10, 6))
Profit_Analysis.plot(kind = "bar", 
title = "Profit Margin by Vehicle Type")
plt.ylabel("Profit Margin (%)")
plt.xticks(rotation = 0)
plt.show()
```

---
Customer Segmentation:-

```python
Customer_Seg = df.groupby("Customer_Type")["Revenue_USD"].sum()

plt.figure(figsize = (10, 6))
Customer_Seg.plot(kind = "pie", autopct = "%1.1f%%")
plt.title("Customer Contribution")
plt.ylabel("")
plt.show()
```

---
Sales Channel Performance:-

```python
Channel_Perf = df.groupby("Sales_Channel")["Revenue_USD"].sum()

plt.figure(figsize = (10, 6))
Channel_Perf.plot(kind = "bar",
title = "Sales Channel Performance")
plt.ylabel("Revenue")
plt.xlabel("Sales Channel")
plt.xticks(rotation = 0)
plt.show()
```

---
Discount Impact By Vehicle Type:-

```python
Discount_Vehicle = df.groupby(["Discount_Band", "Vehicle_Type"], observed = True)["Profit_Margin_%"].mean().unstack()

plt.figure(figsize = (10, 6))
Discount_Vehicle.plot(kind = "bar", stacked = True)
plt.title("Discount Impact By Vehicle Type")
plt.ylabel("Avg. Profit Margin (%)")
plt.xlabel("Discount Band %")
plt.xticks(rotation = 0)
plt.legend(loc = "lower center", bbox_to_anchor = (0.5, -0.3), ncol = 6)
plt.show()
```
