# Advanced Analytics

Objectives:-
1. Identify high-value vs low-value customers.
2. Support customer strategy.

---
Import Libraries:-

```python
import pandas as pd
import numpy as np
```

---
Load Clean Data:-

```python
file_path = ('../Data/02_Clean_Data/MM_Clean_Data.csv')
df = pd.read_csv(file_path)

df["Date"] = pd.to_datetime(df["Date"])
df.head()
```

---
Data Preparation for Advanced Analysis:-

```python
df_advanced = df[[
    "Region", 
    "Customer_Type", 
    "Revenue_USD", 
    "Profit_Margin_%", 
    "Discount_%"
]].copy()
```

---
Part 1: Customer Segmentation:


Apply KMeans Clustering:-

```python
from sklearn.cluster import KMeans

X = df_advanced[["Revenue_USD", "Profit_Margin_%"]]

kmeans = KMeans(n_clusters=3, random_state = 42)
df_advanced["Cluster"] = kmeans.fit_predict(X)
```

---
Analyze Cluster Characteristics:-

```python
Cluster_Summary = df_advanced.groupby("Cluster").agg({
    "Revenue_USD": "mean",
    "Profit_Margin_%": "mean",
    "Customer_Type": "count"
}).rename(columns={"Customer_Type": "Customer_Count"})

Cluster_Summary
```

---
Part 2: Market Expansion Analysis:

Region-wise Performance:-

```python
Region_Analysis = df.groupby("Region").agg({
    "Revenue_USD": "sum",
    "Profit_USD": "sum",
    "Units_Sold": "sum"
})

Region_Analysis
```

---
Identify High Growth Markets:-

```python
Region_Analysis["Profit_Margin_%"] = (
    Region_Analysis["Profit_USD"] / Region_Analysis["Revenue_USD"]
) * 100

Region_Analysis.sort_values(by = "Revenue_USD", ascending = False)
```

---
Part 3: Scenario Analysis (Customer Strategy):

Scenario 1: Discount Increase Impact:-

```python
df["New_Revenue"] = df["Revenue_USD"] * 1.05
```

---
Compare Old vs New Revenue:-

```python
Comparison = df[["Revenue_USD", "New_Revenue"]].sum()
Comparison
```

---
Scenario 2: Profit Impact:-

```python
df["New_Profit"] = df["New_Revenue"] - df["Cost_USD"]
Profit_Comparison = df[["Profit_USD", "New_Profit"]].sum()
Profit_Comparison
```
