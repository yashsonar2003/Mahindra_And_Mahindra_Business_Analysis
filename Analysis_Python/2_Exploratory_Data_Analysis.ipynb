# Exploratory Data Analysis (EDA)

Objective:-
1. Revenue Growth
2. Market Expansion


---
Import Libraries:-

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

---
Load Cleaned Data:-

```python
file_path = ('../Data/02_Clean_Data/MM_Clean_Data.csv')
df = pd.read_csv(file_path)


df["Date"] = pd.to_datetime(df["Date"])
df.head()
```

---
Ensure Proper Month Order:-

```python
df["Month"] = df["Date"].dt.month
df["Month_Name"] = df["Date"].dt.strftime("%b")

month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
```

---
1. Sales Trend Over Time (Revenue Growth):-

```python
Monthly_Sales = df.groupby("Month_Name")["Revenue_USD"].sum().reindex(month_order)

Monthly_Sales.plot(marker= "o")
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

---
2. Region-Wise Performance (Market Expansion):-

```python
Region_Sales = df.groupby("Region")["Revenue_USD"].sum().sort_values(ascending = False)

plt.figure(figsize = (10, 6))
Region_Sales.plot(kind = "bar")
plt.title("Revenue By Region")
plt.xlabel("Region")
plt.ylabel("Revenue")
plt.xticks(rotation = 0)
plt.show()
```

---
3. Vehicle Segment Comparison:-

```python
Vehicle_Sales = df.groupby("Vehicle_Type")["Revenue_USD"].sum().sort_values(ascending = False)

plt.figure(figsize = (10, 6))
Vehicle_Sales.plot(kind = "bar")
plt.title("Vehicle Performance")
plt.xlabel("Vehicle Type")
plt.ylabel("Revenue")
plt.xticks(rotation = 0)
plt.show()
```

---
4. Correlation Heatmap:-


```python
corr = df[["Revenue_USD","Cost_USD","Profit_USD","Units_Sold","Discount_%"]].corr()

plt.figure(figsize = (10, 6))
sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.xticks(rotation = 0)
plt.yticks(rotation = 0)
plt.show()
```
