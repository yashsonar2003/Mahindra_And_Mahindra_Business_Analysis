# Data Cleaning & Preparation

Objective:-
- Data reliability (supports ALL 5 business objectives)

---
Import & Load Data:-

```python
import pandas as pd
import numpy as np

file_path = ('../Data/01_Raw_Data/MM_Raw_Data.csv')
df = pd.read_csv(file_path)

df.head()
```

---
Data Overview:-

```python
df.info()
df.describe()
df.isnull().sum()
```

---
Missing Values Handling:-

```python
df["Customer_Rating"] = df["Customer_Rating"].fillna(df["Customer_Rating"].mean())
```

---
Data Type Conversion:-

```python
df["Date"] = pd.to_datetime(df["Date"], format = "%d-%m-%Y")
```

---
Outlier Treatment (IQR Method):-

```python
Q1 = df["Revenue_USD"].quantile(0.25)
Q3 = df["Revenue_USD"].quantile(0.75)
IQR = Q3 - Q1

df = df[(df["Revenue_USD"] >= Q1 - 1.5*IQR) & 
        (df["Revenue_USD"] <= Q3 + 1.5*IQR)]
```

---
Feature Engineering:-

```python
df["Profit_Margin_%"] = (df["Profit_USD"] / df["Revenue_USD"]) * 100
```

---
Date Features:-

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Month_Name"] = df["Date"].dt.strftime("%b")
```

---
Save Clean Data:-

```python
Cleaned_Path = ('../Data/02_Clean_Data/MM_Clean_Data.csv') 
df.to_csv(Cleaned_Path, index = False)

print("Cleaned Dataset Saved Successfully")
```
